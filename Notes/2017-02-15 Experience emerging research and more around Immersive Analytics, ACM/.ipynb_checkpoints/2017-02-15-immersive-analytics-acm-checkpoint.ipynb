# Experience Emerging Research and More Around Immersive Analytics

**Todd Margolis, Product Manager, Qlik**

Agenda 

6:30 Doors Open, Food & Networking   
7:00 Presentation   

**Event Details**

Interested in learning more about Immersive Analytics?  Join Todd Margolis from Qlik as he discusses this area of emerging research  that blends analytical reasoning with immersive virtual spaces to enhance collaborative decision support. Through the use of new technologies such as virtual reality, augmented reality and large scale tiled display environments, Immersive Analytics can be used to connect colleagues both synchronously and asynchronously as well as locally and remotely.  Examples will be shown for healthcare, manufacturing and competitive intelligence demonstrated by the Qlik analytics platform with Oculus Rift, Microsoft Hololens through Unity 3D and the SAGE2 collaboration middleware.

**Speaker Bio**

Todd Margolis is a transmedia producer of immersive platforms. He is currently a Product Manager at Qlik where his focus on Collaboration enhances the potential for team analytics.

His research investigates how emerging media technologies enable new forms of human and computer interactions. His work draws on 20+ years of experience creating tele-collaborative immersive and interactive artworks and systems. He has published numerous papers on mixed reality art and interaction and lectures internationally on Immersive Analytics. His work has been shown in museums, festivals and galleries around the world.
