## TF tensor manipulations
Tensor flow provides tools for manipulating tensors in the same ways that you can manipulate Numpy arrays.  For example here's the tensor flow version of the slice operation.  
https://www.tensorflow.org/versions/r0.9/api_docs/python/array_ops.html#slice
The inputs for the tf.slice() function are the tensor you want to slice, the beginning index tuple and the size of the slice in each dimension.  You can also name the slice.  

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline


"""
This is a 3 dimensional array, tensor if you are fancy. A tensor is just a fancy
of way saying multi-dimensional array because some people think of array as only 
being 1-dimensional. Another way of saying dimension is rank, so a 3-dimensional
array is the same as a rank 3 tensor.
TensorFlow because your tensors flow through the code?

The left most number is 4, so we have 4 of 2-dimensional arrays.
The middle number is 2, so we have in each 2-dimenional array of length 2.
The right number is 3, so each 1-dimensional array (i.e. each row) is 3 long.
"""
fake_data = np.array(range(24)).reshape((4, 2, 3))
print(fake_data) # take a look at fake_data

```

```python
# tf.slice is in essence cropping a picture by "click" and "drag" 

"""
Colors pictures are often represented by 3-dimensional arrays.
The left number is None; it is an arbitrary number for the height of picture
The middle number is None; it is an arbitrary number for the width of picture
The right number is 3, for the 3 colors of the picture (Red-Green-Blue)
"""
image_template = tf.placeholder("uint8", [None, None, 3])

slice = tf.slice(image_template, [1, 0, 0], [2, -1, -1])
"""
In this slice, the left array is the starting point of the slice. It's like clicking a point
In the right array, it's how many elements to keep. -1 means all. It's like dragging
Hence, start at the 1st point of the 0th index,
                      0th point of the 1st index,
                      0th point in the 2nd index.
Keep 2 points in the 0th index,
     all points in 1th index,
     all points in 2nd index.
"""

with tf.Session() as sess:
    result = sess.run(slice, feed_dict={image_template: fake_data})
    # all we need is to run this to get the results; other stuff can happen outside of the session

print(result.shape)
print("\n")
print(result)
    # notice, we started with index 1 array
```

```python
# Suppose you didn't want to hard code, and you want to split something in half

height, width, depth = fake_data.shape
slice = tf.slice(image_template, [1, 0, 0], [height // 2, -1, -1])

with tf.Session() as sess:
    result = sess.run(slice, feed_dict={image_template: fake_data})
    print(result.shape)
    print(result)
    # notice, we started with index 1 array
```

```python
print neural_net_pic.shape # makes sense as the picture is 722 pixels tall, 600 pixels wide.
# The 3 is for number of colors: Red, Green, Blue
```

```python
image = tf.placeholder("uint8", [None, None, 3])
slice = tf.slice(image, [200, 210, 0], [350, 300, -1])
"""
Here, the first array is the first point where basically I'm clicking to crop to an image
Start at the 200th row, 210th column, 0th color.
The second array is how much stuff to keep--this is the dragging part.
Keep 350 rows, 300 columns, all the colors
"""

with tf.Session() as sess:
    sliced_pic = sess.run(slice, feed_dict={image: neural_net_pic})
    print(sliced_pic.shape)

plt.figure(num=None, figsize=(12, 5))
plt.imshow(sliced_pic)
```

### Splitting Tensors

```python
# Is there a way to SPLIT things instead of manually slicing things?
# tf.split is a way to split into equally sized pieces. Whew!

boring_matrix = np.arange(5 * 30).reshape(5, 30)
print(boring_matrix)

which_index_to_split = 1
how_many_pieces_you_want_to_get_out = 3
split0, split1, split2 = tf.split(which_index_to_split, how_many_pieces_you_want_to_get_out, boring_matrix)

print('\n')
print(split0) # doesn't print what you want to see
print('\n')

with tf.Session() as sess:
    print sess.run(split0)
# As promised, it splits on index 1 which is the columns. You keep all the rows.
# Now you have 1/3 the columns.
```
